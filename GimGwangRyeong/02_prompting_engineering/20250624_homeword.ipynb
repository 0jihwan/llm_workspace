```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
```

```python
from openai import OpenAI
client = OpenAI(
    api_key=OPENAI_API_KEY,
)
```

```python
system_message = """
- LLM이 영어 텍스트에서 학습 가치가 있는 주요 단어 또는 어구를 추출하여 단어장으로 만들어야 한다.
- 너무 쉬운 단어는 제외하고, 고유명사, 전문용어, 관용구 등도 포함한다.
- 각 카드에는 (1)단어/어구, (2)뜻(영어 또는 한글), (3)품사, (4)예문(영어)을 포함한다.
- 각 카드는 Anki에서 바로 쓸 수 있는 포맷(각 항목은 세미콜론 ; 으로 구분, 한 줄에 한 카드)으로 출력한다.

### Steps ###
1. 문장을 읽고 사람들이 많이 쓰거나 쉬운 단어, 고유명사, 전문용어, 관용구 등은 제외한다.
2. 남아있는 단어를 뜻을 영어로 표현하고 그 단어의 품사와 예문을 만든다.
3. 그리고 그 내용들을 '단어/어구;뜻(영어 또는 한글);품사;예문(영어)' 형식으로 정리한다.
4. 모든 단어를 위처럼 표현했으면 출력한다.

### Output Format ###
제공된 문장을 바탕으로 단어장을 만드세요.

- 예시 문장 : 예시로 들어갈 문장
- 출력 내용 : 단어/어구;뜻(영어 또는 한글);품사;예문(영어) ...

### Examples ###
- 예시 문장 : The startup aims to revolutionize the renewable energy sector by leveraging cutting-edge battery technology.
- 출력 내용 :
revolutionize;to completely change something for the better;verb;The startup aims to revolutionize the renewable energy sector.
renewable energy;energy from sources that are naturally replenished;noun;Renewable energy is vital for a sustainable future.
leverage;to use something to maximum advantage;verb;They plan to leverage battery technology to gain a competitive edge.
cutting-edge;very advanced and modern;adjective;The company develops cutting-edge battery solutions.
battery technology;technology related to the design and use of batteries;noun;Cutting-edge battery technology is crucial in electric vehicles.

### Extra Instructions ###
- 단어장에 들어갈 단어는 전체적으로 사람들이 잘 모르는 걸로 해야 한다.
- 단어는 하나씩만 들어가야 한다.

"""
sentence = 'The committee will convene next week to discuss the new policy proposals and reach a consensus.'

user_message = f"""
다음 문장으로 단어장을 만드세요.
{sentence}
"""

response = client.chat.completions.create(
  model="gpt-4.1",
  messages=[
    {
      "role": "system",
      "content": [
        {
          "type": "text",
          "text": system_message
        }
      ]
    },
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": user_message
        }
      ]
    },
  ],
  response_format={
    "type": "text"
  },
  temperature=1,
  max_completion_tokens=2048,
  top_p=1,
  frequency_penalty=0,
  presence_penalty=0
)
print(response.choices[0].message.content)
```
